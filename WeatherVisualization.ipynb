```python
import csv
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
filename = "data.csv"


max_temp=list()
min_temp=list()
date_ = list()

with open(filename) as opened_file:
    reade_file = csv.reader(opened_file)
    next(reade_file)


# looping throu all rows and getting the Max TemperatureF
    for row in reade_file:
        occurance = datetime.strptime(row[0],'%Y-%m-%d')
        item = int(row[1])
        mininum = int(row[3])
        
        date_.append(occurance)
        max_temp.append(item)
        min_temp.append(mininum)
        

# plotting
fig = plt.figure(dpi=100, figsize=(10, 6))
plt.plot(date_,max_temp,c="green")
plt.plot(date_,min_temp, c="orange")

# fill_between method fills and colors the space between the to plots
plt.fill_between(date_,max_temp,min_temp, facecolor="blue",alpha=.2)

# labeling
plt.title("Daily high & low temperatures - 2014", fontsize=10)
plt.xlabel("date")
plt.ylabel("temp")

# The call to fig.autofmt_xdate()
# at x draws the date labels diagonally to prevent them from overlapping. 
fig.autofmt_xdate()
plt.show()
```

dealing with missing data points.
the data stored in the death_valley has soe missing points. when we run the program, we encounter an error. in the bellow cell, we will fix the issue.
we do so by adding a try except block or by adding a continue

```python
title="Daily high & low temperatures  2014\nDeath Valley, CA - 2014"
filename = "death_valley_2014.csv"

max_temp=list()
min_temp=list()
date_ = list()

with open(filename) as opened_file:
    reade_file = csv.reader(opened_file)
    next(reade_file)



    for row in reade_file:
#         trapping empty data points
        try:
            occurance = datetime.strptime(row[0],'%Y-%m-%d')
            item = int(row[1])
            mininum = int(row[3])
        except ValueError:
            print(f"missing data for date {occurance}")
        else:
            date_.append(occurance)
            max_temp.append(item)
            min_temp.append(mininum)
        
        
fig = plt.figure(dpi=100, figsize=(10, 6))

plt.plot(date_,max_temp,c="green")
plt.plot(date_,min_temp, c="orange")

# fill_between method fills and colors the space between the to plots
plt.fill_between(date_,max_temp,min_temp, facecolor="blue",alpha=.2)

# labeling
plt.title(title, fontsize=10)
plt.xlabel("date")
plt.ylabel("temp")


fig.autofmt_xdate()
plt.show()
```
